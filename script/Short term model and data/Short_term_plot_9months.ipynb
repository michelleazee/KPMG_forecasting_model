```python
def generatedataframe(df1, df2):
  df1=df1.drop_duplicates(subset='ds')
  df2=df2.drop_duplicates(subset='ds')
  df2=df2.rename(columns={"yhat": "yhat_modified"})
  df2=df2.rename(columns={"yhat_lower": "yhat_lower_modified"})
  df2=df2.rename(columns={"yhat_upper": "yhat_upper_modified"})
  df2=df2.rename(columns={"y": "y_modified"})
  df2=df2.rename(columns={"cutoff": "cutoff_modified"})
  df=pd.merge(df1, df2, on='ds')
  df=df[['ds','yhat','yhat_lower','yhat_upper','y','cutoff','yhat_modified','yhat_lower_modified','yhat_upper_modified','y_modified','cutoff_modified']]
  df.index=range(len(df))
  df['ds']=pd.to_datetime(df['ds'])
  df=df[df['ds']>'2010-01-01']
  return df
```

```python
m1=generatedataframe(one_month,one_month_modified)
m2=generatedataframe(two_month,two_month_modified)
m3=generatedataframe(three_month,three_month_modified)
m4=generatedataframe(four_month,four_month_modified)
m5=generatedataframe(five_month,five_month_modified)
m6=generatedataframe(six_month,six_month_modified)
m7=generatedataframe(seven_month,seven_month_modified)
m8=generatedataframe(eight_month,eight_month_modified)
m9=generatedataframe(nine_month,nine_month_modified)
```

```python
def plotpredictionresults(df, title1, title2):
  plt.figure(figsize=(30,5))
  plt.subplot(1,2,1)  
  plt.plot(df['ds'],df['yhat'],label='yhat')
  plt.plot(df['ds'],df['y'],label='y')
  plt.ylim(-0.2,0.2)
  plt.title(title1)
  plt.xlabel('Date')
  plt.ylabel('Prediction Results')
  plt.legend()
  plt.subplot(1,2,2)  
  plt.plot(df['ds'],df['yhat_modified'],label='yhat_modified')
  plt.plot(df['ds'],df['y_modified'],label='y_modified')
  plt.ylim(-1.5,1.5)
  plt.title(title2)
  plt.xlabel('Date')
  plt.ylabel('Prediction Results')
  plt.legend()
  plt.show()
```

```python
plotpredictionresults(m1,'One Month Predicion', 'One Month Predicion Modified')
```
